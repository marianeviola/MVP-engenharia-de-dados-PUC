```sql
WITH TotalVagas AS (
    SELECT COUNT(DISTINCT ID) AS total_vagas FROM jobs.silver.vagas_skills_limpas
)
SELECT
    T2.SKILL_CATEGORY_NAME,
    COUNT(DISTINCT T1.ID) AS Total_Vagas_Impactadas,
    ROUND(COUNT(DISTINCT T1.ID) * 100.0 / (SELECT total_vagas FROM TotalVagas), 2) AS Porcentagem_Vagas
FROM jobs.gold.fato_demanda T1
JOIN jobs.gold.dim_habilidade T2
    ON T1.ID_HABILIDADE_HASH = T2.ID_HABILIDADE_HASH
GROUP BY 1
ORDER BY Total_Vagas_Impactadas DESC;
```
