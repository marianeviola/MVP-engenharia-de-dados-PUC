```sql
WITH TotalHabilidades AS (
    -- Conta o total de aparições de skills no FATO (demanda total)
    SELECT COUNT(ID) AS total_demandas
    FROM jobs.gold.fato_demanda
),
DemandasSoftware AS (
    -- Conta o total de aparições (demandas) que são software skills
    SELECT COUNT(T1.ID) AS total_software_demandas
    FROM jobs.gold.fato_demanda T1
    JOIN jobs.gold.dim_habilidade T2
        ON T1.ID_HABILIDADE_HASH = T2.ID_HABILIDADE_HASH
    WHERE T2.IS_SOFTWARE = TRUE
)
SELECT
    ROUND((S.total_software_demandas * 100.0 / T.total_demandas), 2) AS Porcentagem_Software_Exigida
FROM DemandasSoftware S
CROSS JOIN TotalHabilidades T;
```
