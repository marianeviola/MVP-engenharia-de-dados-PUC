```sql
%sql
-- Cria o schema bronze dentro do CATÁLOGO 'jobs'
CREATE SCHEMA IF NOT EXISTS jobs.bronze;

-- Cria a tabela BRONZE (dados_brutos), copiando os dados da fonte CSV (jobs.nl_jobs)
CREATE OR REPLACE TABLE jobs.bronze.dados_brutos AS
SELECT 
    ID,
    LAA_COUNTRY,
    SKILL_NAME,
    SKILL_CATEGORY_NAME,
    SKILL_SUBCATEGORY_NAME,
    IS_SOFTWARE
FROM
    nl_jobs;

SELECT 'Camada BRONZE criada em jobs.bronze.dados_brutos (Cópia Exata da Fonte).' AS Status;
```
